```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def plot_pca(X, y):
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    plt.figure()
    plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="tab10")
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("PCA projekcija podataka")
    plt.show()
```
